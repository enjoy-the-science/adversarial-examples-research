```python
import pandas as pd
import os
```

```python
name = ['HGG', 'LGG']
path = 'MICCAI_BraTS_2019_Data_Training'

data_csv = {'Grade': [], 'BraTS_2019_subject_ID': []}

for cls in name:
    dirs = os.listdir(os.path.join(path, cls))
    cls_list = [cls]*len(dirs)
    
    data_csv['Grade'] += cls_list
    data_csv['BraTS_2019_subject_ID'] += dirs

data_csv = pd.DataFrame(data_csv)
```
